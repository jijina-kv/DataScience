```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.decomposition import FactorAnalysis as FA
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
data=load_digits()
data.keys()
```

```python

df1=pd.DataFrame(data['data'],columns=data['feature_names'])
# Scale data before applying PCA
scaling=StandardScaler()
print(data['data'].shape) 
# Use fit and transform method
scaling.fit(df1)
Scaled_data=scaling.transform(df1)
 
# Set the n_components=3
principal=PCA(n_components=3)
principal.fit(Scaled_data)
xPCA=principal.transform(Scaled_data)
# Check the dimensions of data after PCA
print(xPCA.shape)
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize = (10,6))
c_map = plt.cm.get_cmap('jet', 10)
plt.scatter(xPCA[:, 0], xPCA[:, 1], s = 15,
            cmap = c_map , c = data.target)
plt.colorbar()
plt.xlabel('PC-1') , plt.ylabel('PC-2')
plt.show()
```

```python
df1=pd.DataFrame(data['data'],columns=data['feature_names'])
y = data.target
# Scale data before applying PCA
scaling=StandardScaler()
print(data['data'].shape) 
# Use fit and transform method
scaling.fit(df1)
Scaled_data=scaling.transform(df1)
 
# Set the n_components=2
principal=LDA(n_components=2)
principal.fit(Scaled_data,y)
xLDA=principal.transform(Scaled_data)
# Check the dimensions of data after PCA
print(xLDA.shape)
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize = (10,6))
c_map = plt.cm.get_cmap('jet', 10)
plt.scatter(xLDA[:, 0], xLDA[:, 1], s = 15,
            cmap = c_map , c = data.target)
plt.colorbar()
plt.xlabel('LD-1') , plt.ylabel('LD-2')
plt.show()
```
